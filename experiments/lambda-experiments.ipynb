```python
def square_number(x):
    return x**2
```

```python
def print_details(x, msg):
    print(f"{x} {msg}")
```

```python
def print_and_square(x, func):
    return func(square_number(x))
```

```python
square_number(3)
```

```python
print_details(3, "hello world")
```

```python
x = 6
msg = "hello world"

print_and_square(x, lambda x: print_details(x, msg))
```

```python
print_details(3, msg)
```
